```python
#removing the word county 

county_df=pd.read_csv('IL_county_pop.csv')
county_df['County']= [x.split(' County')[0] for x in county_df['County']]
county_dic= pd.Series([ [x, 0, 1, 0] for x in county_df.Population.values], index=county_df.County).to_dict()
```

```python
#making the county dict
county_dic= pd.Series([ [int(x.replace(',','')), 0, 1, 0] for x in county_df.Population.values], index=county_df.County).to_dict()

```

```python
import csv
# w = csv.writer(open("county_dic.csv", "w"))
# for key, val in county_dic.items():
#     w.writerow([key, val])
    
with open('county_dic.csv', 'w', newline='') as csvfile:
    fieldnames = ['county', 'val_list']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for key, val in county_dic.items():
    #writer.writeheader()
        writer.writerow({'county': key, 'val_list': val})
```
